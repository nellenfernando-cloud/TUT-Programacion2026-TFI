```python
import networkx as nx
import matplotlib.pyplot as plt

# Definimos la red
red = {
    "A": ["B", "C"],
    "B": ["C", "D"],
    "C": ["D", "E"],
    "D": ["E", "F"],
    "E": ["F", "G"],
    "F": ["G", "H"],
    "G": ["H", "A"],
    "H": ["A", "B"]
}

# Crear grafo dirigido
G = nx.DiGraph()

# Agregar enlaces
for origen in red:
    for destino in red[origen]:
        G.add_edge(origen, destino)

# Mostrar nodos disponibles
print("Nodos disponibles:")
for nodo in red:
    print(nodo)

# Solicitar nodo
origen = input("Ingrese un nodo: ").upper()

if origen in red:
    print(f"\nEl nodo {origen} puede comunicarse directamente con:")
    for destino in red[origen]:
        print(destino)
else:
    print("Nodo inexistente.")

# Dibujar red
plt.figure(figsize=(8,6))
pos = nx.spring_layout(G, seed=42)

nx.draw(
    G,
    pos,
    with_labels=True,
    node_size=2500,
    arrows=True
)

# Resaltar nodo seleccionado
if origen in red:
    nx.draw_networkx_nodes(
        G,
        pos,
        nodelist=[origen],
        node_size=3000
    )

plt.title("Topología de la Red")
plt.show()
```
